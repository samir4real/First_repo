##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
# Write your code here
print("Square Root of Vector Elements (using for loop):")

vector = np.array([25, 16, 9, 4, 1])
print(f"Original vector: {vector}")

sqrt_results = []
for element in vector:
    sqrt_value = np.sqrt(element)
    sqrt_results.append(float(sqrt_value))

print(f"Square roots: {sqrt_results}")
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
# Write your code here

print("Function to Generate Vector:")

def generate_vector(n):
    vector = []
    for i in range(n, 0, -1):
        vector.append(i ** 2)
    return np.array(vector)

# Test with n=5
result_5 = generate_vector(5)
print(f"Input: 5")
print(f"Output: {result_5}")

# Test with n=10
result_10 = generate_vector(10)
print(f"\nInput: 10")
print(f"Output: {result_10}")
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Write your code here
print("Plotting 3D Surface: x + y + z = 1")

x_range = np.arange(-3, 3.5, 0.5)
y_range = np.arange(-3, 3.5, 0.5)

X, Y = np.meshgrid(x_range, y_range)

Z = 1 - X - Y

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, edgecolor='none')

ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title('3D Surface: x + y + z = 1', fontsize=14, fontweight='bold')

fig.colorbar(surface, shrink=0.5, aspect=5)

plt.show()

print(f"Grid size: {X.shape[0]} x {X.shape[1]} points")
```

```python
# Problem 09: Image Processing with NumPy



image_path = "Lenna_(test_image).png"
image = io.imread(image_path)

print("Image shape (Height, Width, Channels):", image.shape)

min_val = np.min(image)
max_val = np.max(image)
print("Minimum pixel value:", min_val)
print("Maximum pixel value:", max_val)

mean_val = np.mean(image)
std_val = np.std(image)
print("Mean pixel value:", mean_val)
print("Standard deviation:", std_val)

plt.figure(figsize=(4,4))
plt.imshow(image)
plt.title("Original Lena Image")
plt.axis("off")
plt.show()

resized_image = transform.resize(image, (50, 50), anti_aliasing=True)
print("Resized image shape:", resized_image.shape)

gray_image = color.rgb2gray(resized_image)
print("Grayscale image shape:", gray_image.shape)

plt.figure(figsize=(4,4))
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale 50x50 Image")
plt.axis("off")
plt.show()

imsave("lena_grayscale_50x50.png", (gray_image * 255).astype(np.uint8), cmap='gray')
print("Grayscale image saved as 'lena_grayscale_50x50.png'")

vectorized = np.reshape(gray_image, (2500, 1))
print("Vectorized image shape:", vectorized.shape)
print("First 10 elements of vectorized image:\n", vectorized[:10])
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
# Write your code here
print("Taylor Approximation Analysis:")

x_taylor = np.arange(-1, 1.01, 0.01)

T2_e2x = 1 + 2*x_taylor + (2*x_taylor)**2 / 2

T2_e3x = 1 + 3*x_taylor + (3*x_taylor)**2 / 2

T2_e5x = 1 + 5*x_taylor + (5*x_taylor)**2 / 2

method1 = T2_e2x * T2_e3x

method2 = T2_e5x

original = np.exp(5*x_taylor)

mse1 = np.mean((original - method1)**2)
mse2 = np.mean((original - method2)**2)

print(f"\nMethod 1 (e^2x * e^3x Taylor): MSE = {mse1:.6f}")
print(f"Method 2 (e^5x Taylor direct): MSE = {mse2:.6f}")
print(f"\nMore accurate method: Method {'2' if mse2 < mse1 else '1'}")
print(f"MSE difference: {abs(mse1 - mse2):.6f}")

plt.figure(figsize=(14, 6))

plt.plot(x_taylor, original, 'k-', linewidth=2.5, label='Original: e^(5x)', alpha=0.8)
plt.plot(x_taylor, method1, 'r--', linewidth=2, label='Method 1: T2(e^2x) Ã T2(e^3x)', alpha=0.7)
plt.plot(x_taylor, method2, 'b-.', linewidth=2, label='Method 2: T2(e^5x)', alpha=0.7)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Taylor Approximation Comparison: e^(5x)', fontsize=14, fontweight='bold')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, alpha=0.3)
plt.xlim(-1, 1)

textstr = f'MSE Method 1: {mse1:.6f}\nMSE Method 2: {mse2:.6f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

print("\nâ Analysis complete!")
```
